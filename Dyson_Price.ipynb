```python
import requests 
from bs4 import BeautifulSoup
r = requests.get("https://www.price.com.hk/search.php?g=A&q=dyson")
soup = BeautifulSoup(r.text)
```

```python
price=soup.find('div', {'class':'line line-01 price-range'}).find('span').text
price
```

```python
print(price.replace('\n', ''))
```

```python
page=soup.find('div', {'class':'pagination-total'}).text #'共 29 頁'
total_page=int(page[2:4])
total_page
```

```python
for j in range(total_page+1):
    r = requests.get(f"https://www.price.com.hk/search.php?g=A&q=dyson={j}")
    soup = BeautifulSoup(r.text)
    for i in soup.findAll('div',{'class':'line line-01 price-range'}):
        print(i.find('span').text)
```
