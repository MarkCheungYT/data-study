```python
df_flights_total=df_flights.groupby(['AIRLINE','ORIGIN_AIRPORT']).agg({'FLIGHT_NUMBER':'count'})
```

```python
df_flights_total
```

```python
df_flights_operated=df_flights.loc[df_flights['CANCELLED']==0].groupby(['AIRLINE','ORIGIN_AIRPORT']).agg({'FLIGHT_NUMBER':'count'})
```

```python
df_flights_operated
```

```python
df_flights_operated['FLIGHT_NUMBER']
```

```python
df_flights_total["Operated_flight"]=df_flights_operated['FLIGHT_NUMBER']
```

```python
df_flights_total["Operated_rate"]=100*df_flights_total["Operated_flight"]/df_flights_total['FLIGHT_NUMBER']
```

```python
df_flights_total
```

```python
df_flights_delay_and_cancelled=df_flights.loc[(df_flights['ARRIVAL_DELAY']>0) & (df_flights['CANCELLED']==0)].groupby(['AIRLINE','ORIGIN_AIRPORT']).agg({'FLIGHT_NUMBER':'count'})
```

```python
df_flights_delay_and_cancelled
```

```python
df_flights_total['delayed_and_cancelled_flight']=df_flights_delay_and_cancelled['FLIGHT_NUMBER']
```

```python
df_flights_total["delay_rate"]=100*df_flights_total['delayed_and_cancelled_flight']/df_flights_total['FLIGHT_NUMBER']
```

```python
df_flights_total
```

```python
df_flights_average_delay_time=df_flights.loc[(df_flights['ARRIVAL_DELAY']>0) & (df_flights['CANCELLED']==0)].groupby(['AIRLINE','ORIGIN_AIRPORT']).agg({'ARRIVAL_DELAY':'mean'})
```

```python
df_flights_average_delay_time
```

```python
df_flights_total['average_delay_minutes']=df_flights_average_delay_time['ARRIVAL_DELAY']
```
